```python
class Shop:
    def __init__(self, shopName, maxScooters):
        self.shopName = shopName
        self.maxScooters = maxScooters
        self.availableScooters = maxScooters
```

```python
class Customer:
    def __init__(self, customerName, hourlyScooters = None, dailyScooters = None, weeklyScooters = None, currentPrice = None, shopName = None):
        self.customerName = customerName
        self.hourlyScooters = hourlyScooters
        self.dailyScooters = dailyScooters
        self.weeklyScooters = weeklyScooters
        self.currentPrice = currentPrice
        self.shopName = shopName
        
    def checkCustomerInfo(self):
        return f"===================Customer Info =======================\nCustomer Name: {self.customerName}\nHourly Scooters Rented: {self.hourlyScooters}\nDaily Scooters Rented: {self.dailyScooters}\nWeekly Scooters Rented: {self.weeklyScooters}\nthe current bill is: {self.currentPrice}\nfrom shop: {self.shopName}"
```

```python
class Rentals:
    __hourlyRate = 10
    __dailyRate = 20
    __weeklyRate = 50
    __discount = 0.15
    __chargedRate = 0
    
    def __init__(self, shop, customer):
        self.shop = shop
        self.customer = customer
    
    def rentScooter(self, scooterCount, rentDuration, rentType):
        if scooterCount <= 0:
            print("You cannot rent 0 scooters")
            return
        if scooterCount > self.shop.availableScooters:
            print(f"{scooterCount} scooters are not available. Only {self.shop.availableScooters} scooters are available")
            return
        
        if rentType == "hour":
            if self.customer.hourlyScooters is None:
                self.customer.hourlyScooters = 0
                
            self.customer.hourlyScooters += scooterCount
        elif rentType == "day":
            if self.customer.dailyScooters is None:
                self.customer.dailyScooters = 0
                
            self.customer.dailyScooters += scooterCount
        elif rentType == "week":
            if self.customer.weeklyScooters is None:
                self.customer.weeklyScooters = 0
                
            self.customer.weeklyScooters += scooterCount
        
        self.shop.availableScooters -= scooterCount
        totalRent = self.calculateRent(scooterCount, rentDuration, rentType)
        if self.customer.currentPrice is None:
            self.customer.currentPrice = 0
        
        self.customer.currentPrice += totalRent
        return totalRent
    
    def returnScooter(self, scooterCount, rentDuration, rentType):
        if scooterCount == 0:
            print("You cannot return 0 scooters.")
            
        if rentType == "hour":
            if self.customer.hourlyScooters < scooterCount or self.customer.hourlyScooters is None:
                print(f"You have not rented {scooterCount} scooters on hourly basis")
            else:
                self.customer.hourlyScooters -= scooterCount
                self.shop.availableScooters += scooterCount
        if rentType == "day":
            if self.customer.dailyScooters < scooterCount or self.customer.dailyScooters is None:
                print(f"You have not rented {scooterCount} scooters on daily basis")
            else:
                self.customer.dailyScooters -= scooterCount
                self.shop.availableScooters += scooterCount
        if rentType == "week":
            if self.customer.weeklyScooters < scooterCount or self.customer.weeklyScooters is None:
                print(f"You have not rented {scooterCount} scooters on weekly basis")
            else:
                self.customer.weeklyScooters -= scooterCount
                self.shop.availableScooters += scooterCount
                
        return self.generateInvoice(scooterCount, rentDuration, rentType)
                
    def calculateRent(self, scooterCount, rentDuration, rentType):
        rate = 0
        if rentType == "hour":
            rate = self.__hourlyRate * scooterCount * rentDuration
            self.__chargedRate = self.__hourlyRate
        elif rentType == "day":
            rate = self.__dailyRate * scooterCount * rentDuration
            self.__chargedRate = self.__dailyRate
        elif rentType == "week":
            rate = self.__weeklyRate * scooterCount * rentDuration
            self.__chargedRate = self.__weeklyRate
        
        if scooterCount > 2:
            rate = rate * (1-self.__discount)
        
        return rate
    
    def generateInvoice(self, scooterCount, rentDuration, rentType):
        if scooterCount <= 0:
            print("You cannot rent 0 scooters")
            return
        if scooterCount > self.shop.availableScooters:
            print(f"{scooterCount} scooters are not available. Only {self.shop.availableScooters} scooters are available")
            return
        
        rent = self.calculateRent(scooterCount, rentDuration, rentType);
        val = f"===================Billing Info =======================\nCustomer Name: {self.customer.customerName}\nNumber of scooter rented: {scooterCount}\nCharged Rate: {self.__chargedRate}\nTotal Rent: {rent}"
        return val
```
